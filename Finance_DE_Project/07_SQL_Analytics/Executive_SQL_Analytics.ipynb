```python
%sql
--Top 10 Customers by Spending

CREATE OR REPLACE TABLE dashboard_top_customers AS
SELECT
  customer_id,
  name,
  country,
  total_spending,
  total_transactions
FROM gold_customer_spending
ORDER BY total_spending DESC
LIMIT 10;

SELECT * FROM dashboard_top_customers;
```

```python
%sql
--Country Revenue Dashboard
CREATE OR REPLACE TABLE dashboard_country_revenue AS
SELECT
    country,
    country_total_amount,
    country_total_transactions
FROM gold_country_analytics
ORDER BY country_total_amount DESC;

SELECT * FROM dashboard_country_revenue;
```

```python
%sql
--Transaction Type Dashboard
CREATE OR REPLACE TABLE dashboard_transaction_type AS
SELECT
    transaction_type,
    total_amount,
    transaction_count
FROM gold_transaction_analytics
ORDER BY total_amount DESC;

SELECT * FROM dashboard_transaction_type;
```

```python
%sql
--High-Risk Customers
CREATE OR REPLACE TABLE dashboard_high_risk_customers AS
SELECT *
FROM gold_customer_risk
WHERE customer_risk_level = 'HIGH';

SELECT * FROM dashboard_high_risk_customers;
```

```python
%sql
--Fraud Summary Dashboard
CREATE OR REPLACE TABLE dashboard_fraud_summary AS
SELECT *
FROM gold_fraud_dashboard;

SELECT * FROM dashboard_fraud_summary;
```

```python
%sql
--Executive KPI Table
CREATE OR REPLACE TABLE executive_kpis AS
SELECT
    COUNT(DISTINCT customer_id) AS total_customers,
    COUNT(*) AS total_transactions,
    ROUND(SUM(transaction_amount),2) AS total_transaction_amount,
    ROUND(AVG(transaction_amount),2) AS avg_transaction_amount,
    SUM(is_fraud) AS fraud_transactions
FROM silver_finance;

SELECT * FROM executive_kpis;
```

```python
%sql
--Monthly Revenue Trend
CREATE OR REPLACE TABLE dashboard_monthly_revenue AS
SELECT
    YEAR(transaction_time) AS year,
    MONTH(transaction_time) AS month,
    ROUND(SUM(transaction_amount),2) AS revenue
FROM silver_finance
GROUP BY
    YEAR(transaction_time),
    MONTH(transaction_time)
ORDER BY year, month;

SELECT * FROM dashboard_monthly_revenue;
```

```python
%sql
--Top Fraudulent Transaction Types
CREATE OR REPLACE TABLE dashboard_fraud_by_type AS
SELECT
    transaction_type,
    COUNT(*) AS fraud_count,
    ROUND(SUM(transaction_amount),2) AS fraud_amount
FROM silver_finance
WHERE is_fraud = 1
GROUP BY transaction_type
ORDER BY fraud_count DESC;

SELECT * FROM dashboard_fraud_by_type;
```
