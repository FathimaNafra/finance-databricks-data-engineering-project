```python
#Read_Raw_Tables
customer_raw_df = spark.table("finance_project_customers")
transaction_raw_df = spark.table("transactions")
```

```python
#Add Audit Columns
from pyspark.sql.functions import current_timestamp

bronze_customers = customer_raw_df.withColumn(
    "ingeshion_time",
    current_timestamp()
)

bronze_transactions=transaction_raw_df.withColumn(
    "ingeshion_time",
    current_timestamp()
)
```

```python
#Create Bronze Tables

bronze_customers.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_customers")

bronze_transactions.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_transactions")
```

```python
#Verify Customers

display(
    spark.sql("""
    SELECT * FROM bronze_customers
    LIMIT 10
    """)
)
```

```python
#Verify Transactions

display(
    spark.sql("""
    SELECT * FROM bronze_transactions
    LIMIT 10
    """)
)
```
