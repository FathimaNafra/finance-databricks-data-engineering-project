```python
from faker import Faker
import pandas as pd
import random

fake=Faker()
customers=[]

for i in range(1,5001):
    customers.append({
        "customer_id":i,
        "name":fake.name(),
        "email":fake.email(),
        "city":fake.city(),
        "country":fake.country(),
        "account_type":random.choice(["Savings","Current"]),
        "created_date":fake.date_between(start_date='-5y',end_date='today')
    })

customer_df=pd.DataFrame(customers)
display(customer_df)
```

```python
spark_customer_df = spark.createDataFrame(customer_df)
spark_customer_df.write\
    .mode("overwrite")\
    .format("delta")\
    .saveAsTable("finance_project_customers")
```

```python
transactions=[]

for i in range(1,100001):
    transactions.append({
        "transaction_id":i,
        "customer_id":random.randint(1,5000),
        "transaction_amount":round(random.uniform(10,5000),2),
        "transaction_type":random.choice([
            "UPI",
            "NEFT",
            "RIGS",
            "Credit Card",
            "Debit Card"
        ]),
        "merchant":fake.company(),
        "transction_city":fake.city(),
        "transaction_time":fake.date_time_this_year(),
        "is_fraud":random.choice([0,0,0,0,0,1])
    })

transaction_df=pd.DataFrame(transactions)
display(transaction_df)
```

```python
spark_transaction_df=spark.createDataFrame(transaction_df)

spark_transaction_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("transactions")
```
